```python
# Import libraries to use to start cleaning the EDD dataset
import csv
import pandas as pd
import numpy as np
import datetime 
```

```python
# Create a path to data file
edd_path = "../../Extract/rawdata/employment_dataset.csv"

# Then read the EDD data and its study results
edd_df = pd.read_csv(edd_path)
edd_df
```

```python
# Before cleaning the data rename area_name to county (aligns with other datasets) using the .rename function 
edd_df.rename(columns = {"area_name": "county"},  
          inplace = True) 
edd_df
```

```python
# Extract only the COVID 19 years which is from Jan 2020 to January 2021 using .loc
covid_year_df = edd_df.loc[edd_df["year"]>=2020]
covid_year_df
```

```python
# Search the types of rows this dataset has & filter according to 'county' (our main indicator)
covid_year_df["area_type"].unique()
```

```python
# Extract only county from the covid_year_df
county_df = covid_year_df.loc[covid_year_df["area_type"]=="County"]
county_df
```

```python
# Clean each county: original (x "" County) to reformatted (x)
county_df["county"] = county_df["county"].str.strip("County")
county_df
```

```python
# LA shows up twice because of seasonally_adjusted_y_n. We only want the Los Angeles NOT seasonally adjusted from the dataset.
LA_filtered_county = county_df.loc[county_df["seasonally_adjusted_y_n"] == "N"]
LA_filtered_county
```

```python
# Drop unneccessary columns: "Unnamed: 0", "area_type","year","month", "status_preliminary_final", "seasonally_adjusted_y_n"
cleaned_covid_year_df = LA_filtered_county.drop(columns= ["Unnamed: 0", "year","month", "area_type", "seasonally_adjusted_y_n","status_preliminary_final"], axis = 1)
cleaned_covid_year_df
```

```python
# Format the date column to yyyy-mm-dd
cleaned_covid_year_df["date"] = cleaned_covid_year_df["date"].apply(pd.to_datetime)
cleaned_covid_year_df.head(100)
```

```python
# Create a new csv without the index
cleaned_covid_year_df.to_csv("../Output/Cleaned Datasets/EDD_cleaned_dataset.csv", index=False)
```

```python
# Import libraries to use to start cleaning the COVID csv
import pandas as pd
import pathlib
```

### Dateset 1: COVID 19 Hospital by County - Cleaning

```python
# Create a path to data file
covidhospitalbycounty_path = "../../Extract/rawdata/covid19hospitalbycounty.csv"

# Then read the COVID data and its study results
covidhospitalbycounty_df = pd.read_csv(covidhospitalbycounty_path)
covidhospitalbycounty_df
```

```python
# Rename the columns for easier reading purposes
covidhospitalbycounty_df.rename(
    columns = {
        'hospitalized_covid_confirmed_patients': 'confirmed_patients'
        ,'hospitalized_suspected_covid_patients':'suspected_patients'
        ,'todays_date':'date'
        ,"icu_suspected_covid_patients" : "icu_covid_suspected"
        ,"icu_covid_confirmed_patients" :"icu_covid_confirmed"}
        ,inplace=True)

covidhospitalbycounty_df.head()
```

```python
# Finding the total number of hospitalized covid patients
covidhospitalbycounty_df["total_covid_patients"] = covidhospitalbycounty_df["suspected_patients"] + covidhospitalbycounty_df["confirmed_patients"] 
covidhospitalbycounty_df
```

```python
#Fill in zero for all NaN using .fillna function
covidhospitalbycounty_df.fillna(0, inplace = True, downcast = 'infer')
covidhospitalbycounty_df
```

```python
# Delete extraneous columns: "hospitalized_covid_patients" and "icu_covid_confirmed_patients"
covidhospitalbycounty_df =covidhospitalbycounty_df.drop(columns= ["hospitalized_covid_patients"])    
covidhospitalbycounty_df.head()
```

```python
#final df for covidhospitalbycounty data
covidhospitalbycounty_df["date"] = covidhospitalbycounty_df["date"].apply(pd.to_datetime)
covidhospitalbycounty_df
```

```python
# This is the final cleaned hospital by county data
covidhospitalbycounty_df.to_csv("../Cleaned_datasets/covidhospitalbycounty_df.csv", index=False)
```

### Dateset 2: COVID 19 Cases - Cleaning

```python
# Create a path to data file
covid19cases_path = "../../Extract/rawdata/covid19cases_test.csv"

# Then read the COVID data and its study results
covid19cases_df = pd.read_csv(covid19cases_path)
covid19cases_df
```

```python
# Rename area to county to be in unison with the other datasets
covid19cases_df.rename(columns = {'area': 'county'},
          inplace=True)
covid19cases_df
```

```python
# Remove "Out of state", "County", "California" row using multiple locs with:  df['a'] > 1) & (df['b'] > 0)) 
covid19cases_df = covid19cases_df.loc[(covid19cases_df['county']!= 'Out of state') & (covid19cases_df['county']!= 'California') & ( covid19cases_df['county']!= 'Unknown')]
covid19cases_df
```

```python
# Delete extraneous columns 
covid19cases_df =covid19cases_df.drop(columns= ["area_type", "cumulative_cases", "cumulative_deaths", "cumulative_total_tests", "cumulative_positive_tests", "cumulative_reported_cases", "cumulative_reported_deaths", "cumulative_reported_tests"])    
covid19cases_df.head()
```

```python
covid19cases_df.fillna(0, inplace = True, downcast = 'infer')
covid19cases_df
```

```python
# Final Dataframe for COVID 19 cases change the format of the data
covid19cases_df["date"] = covid19cases_df["date"].apply(pd.to_datetime)
covid19cases_df
```

```python
# This is the final COVID19 test cases data. Remove index. 
covid19cases_df.to_csv("../Cleaned_datasets/covid19cases_df.csv", index=False)
```

### Dateset 3: COVID 19 Vaccines - Cleaning

```python
# Set path and read a for file
vaccines_path = "../../Extract/rawdata/covid19vaccinesbycounty .csv"
vaccines_df = pd.read_csv(vaccines_path)
vaccines_df
```

```python
# Rename columns by county_name to county and administered_date to date
vaccines_df.rename(columns = {'administered_date':'date'}, 
          inplace=True)

vaccines_df.head()
```

```python
# See unique names for county
vaccines_df['county'].unique()
```

```python
# Remove the following list:
    #('California', 'Outside California', 'Unknown','Oxford', 'Skagit', 'Cottonwood', 'Tarrant','Windham', 'Clallam', 'Snohomish', 'Morgan', 'Hamilton', 'La Paz'
    # , 'Walla Walla', 'Bennington', 'Jackson', 'Carson City', 'Hawaii','Linn', 'Lewis', 'Meeker', 'Kenai Peninsula Borough', 'Kitsap','Maricopa', 'Yavapai', 'Clark')
filtered_counties = vaccines_df.loc[
    (vaccines_df['county']!= 'Outside California') 
    & (vaccines_df['county']!= 'Unknown')
    & (vaccines_df['county']!= 'Oxford')
    & (vaccines_df['county']!= 'Skagit') 
    & (vaccines_df['county']!= 'Cottonwood')
    & (vaccines_df['county']!= 'Tarrant')
    & (vaccines_df['county']!= 'Windham')
    & (vaccines_df['county']!= 'Clallam')
    & (vaccines_df['county']!= 'Snohomish')
    & (vaccines_df['county']!= 'Morgan')
    & (vaccines_df['county']!= 'Hamilton')
    & (vaccines_df['county']!= 'La Paz')
    & (vaccines_df['county']!= 'Walla Walla')
    & (vaccines_df['county']!= 'Bennington')
    & (vaccines_df['county']!= 'Jackson')
    & (vaccines_df['county']!= 'Carson City')
    & (vaccines_df['county']!= 'Hawaii')
    & (vaccines_df['county']!= 'Linn')
    & (vaccines_df['county']!= 'Lewis')
    & (vaccines_df['county']!= 'Meeker')
    & (vaccines_df['county']!= 'Kenai Peninsula Borough')
    & (vaccines_df['county']!= 'Kitsap')
    & (vaccines_df['county']!= 'Maricopa')
    & (vaccines_df['county']!= 'Yavapai')
    & (vaccines_df['county']!= 'Clark')
]

filtered_counties
```

```python
# Drop "calfifornia_flag because this dataset only looks at calforinia 
vaccines = filtered_counties.drop(columns= ["california_flag"])
vaccines
```

```python
# Fill 0 for NaN
vaccines.fillna(0, inplace = True, downcast = 'infer')
vaccines
```

```python
# Final dataset for vaccines with changing the format of the date
vaccines["date"] = vaccines["date"].apply(pd.to_datetime)
vaccines
```

```python
# This is the final vaccine data that will be used in SQL
vaccines.to_csv("../Cleaned_datasets/vaccines.csv", index=False)
```

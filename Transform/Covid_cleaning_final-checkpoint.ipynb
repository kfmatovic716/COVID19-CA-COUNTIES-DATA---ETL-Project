```python
# Libraries imported
import pandas as pd
import pathlib
```

```python
# Set path for file
csvpath = pathlib.Path("Resources/covid19hospitalbycounty.csv")
print(csvpath)
```

```python
# Reading csv 
covidhospitalbycounty_df = pd.read_csv(csvpath)
covidhospitalbycounty_df.head()
```

```python
# Finding the total number of hospitalized covid patients
covidhospitalbycounty_df["total_covid_patients"] = covidhospitalbycounty_df["hospitalized_suspected_covid_patients"] + covidhospitalbycounty_df["hospitalized_covid_confirmed_patients"] 
covidhospitalbycounty_df
```

```python
#Fill in zero for all NaN using .fillna function
covidhospitalbycounty_df.fillna(0, inplace = True, downcast = 'infer')
covidhospitalbycounty_df
```

```python
# Delete extraneous columns: "hospitalized_covid_patients" and "icu_covid_confirmed_patients"
covidhospitalbycounty_df =covidhospitalbycounty_df.drop(columns= ["hospitalized_covid_patients"])    
covidhospitalbycounty_df.head()
```

```python
# Rename the column todays_date':'date'
covidhospitalbycounty_df.rename(columns = {'hospitalized_covid_confirmed_patients': 'confirmed_patients','hospitalized_suspected_covid_patients':'suspected_patients','todays_date':'date',"icu_suspected_covid_patients" : "icu_covid_suspected","icu_covid_confirmed_patients" :"icu_covid_confirmed"},
          inplace=True)

covidhospitalbycounty_df.head()
```

```python
#final df for covidhospitalbycounty data
covidhospitalbycounty_df["date"] = covidhospitalbycounty_df["date"].apply(pd.to_datetime)
covidhospitalbycounty_df
```

```python
# Set path for file
csvpath = pathlib.Path("Resources/covid19cases_test.csv")
print(csvpath)
```

```python
# Read the path
covid19cases_df = pd.read_csv(csvpath)
covid19cases_df.head()
```

```python
# Delete extraneous column
covid19cases_df =covid19cases_df.drop(columns= ["area_type"])    
covid19cases_df.head()
```

```python
covid19cases_df.rename(columns = {'area': 'county'},
          inplace=True)

covid19cases_df.head()
```

```python
# remove out of state row
covid19cases_df = covid19cases_df.loc[( covid19cases_df['area']!= 'Out of state' )]
covid19cases_df
```

```python
# remove unknown row
covid19cases_df = covid19cases_df.loc[( covid19cases_df['area']!= 'Unknown'  ),]
covid19cases_df
```

```python
covid19cases_df.fillna(0, inplace = True, downcast = 'infer')
covid19cases_df
```

```python
#final df for covid19 data
covid19cases_df["date"] = covid19cases_df["date"].apply(pd.to_datetime)
covid19cases_df
```

```python
# Set path for file
csvpath= pathlib.Path("Resources/covid19vaccinesbycounty.csv")
```

```python
vaccines_df = pd.read_csv(csvpath)
vaccines_df.head()
```

```python
# Delete extraneous column
vaccines_df=vaccines_df.drop(columns= ["california_flag"])    
vaccines_df.head()
```

```python
# Fill 0 for NaN
vaccines_df.fillna(0, inplace = True, downcast = 'infer')
vaccines_df
```

```python
# Rename columns by county_name to county and administered_date to date
vaccines_df.rename(columns = {'county_name':'county','administered_date':'date'}, 
          inplace=True)

vaccines_df.head()
```

```python
# Final dataset for vaccines with changing the format of the date
vaccines_df["date"] = vaccines_df["date"].apply(pd.to_datetime)
vaccines_df
```

```python
# This is the final vaccine data that will be used in SQL
vaccines_df.to_csv("vaccines_df.csv")
```

```python
# This is the final hospital by county data
covidhospitalbycounty_df.to_csv("covidhospitalbycounty_df.csv")
```

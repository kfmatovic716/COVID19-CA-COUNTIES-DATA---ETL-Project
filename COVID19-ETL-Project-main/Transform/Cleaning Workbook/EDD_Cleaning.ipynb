```python
import csv
import pandas as pd
import numpy as np
import datetime 
```

```python
# Create path to data files
edd_path = "../../Extract/rawdata/employment_dataset.csv"

# Read the EDD data and the study results
edd_df = pd.read_csv(edd_path)
edd_df
```

```python
# Recreate a new df with only from Jan 2020 to January 2021
covid_year_df = edd_df.loc[edd_df["year"]>=2020]
covid_year_df
```

```python
# Find the types of rows for area_type, aree
covid_year_df["area_type"].unique()
```

```python
# Grab only County from area_type
county_df = covid_year_df.loc[covid_year_df["area_type"]=="County"]
county_df
```

```python
# Remove the county from each name
county_df["area_name"] = county_df["area_name"].str.replace(" County", "")
county_df
```

```python
# Filtered LA county and seasonally_adjusted_y because of duplication 
LA_filtered_county = county_df.loc[county_df["seasonally_adjusted_y_n"] == "N"]
LA_filtered_county
```

```python
# Drop unneccessary columns: "Unnamed: 0", "area_type", "status_preliminary_final", "dates"
cleaned_covid_year_df = LA_filtered_county.drop(columns= ["Unnamed: 0", "year","month", "area_type", "seasonally_adjusted_y_n","status_preliminary_final"], axis = 1)
cleaned_covid_year_df
```

```python
# format date column to mm-dd-yyyy
cleaned_covid_year_df["date"] = cleaned_covid_year_df["date"].apply(pd.to_datetime)
cleaned_covid_year_df.head(100)
```

```python
cleaned_covid_year_df.rename(columns = {"area_name": "county"},  
          inplace = True) 
cleaned_covid_year_df
```

```python
# Delete the automatic index 
# cleaned_covid_year_df.style.hide_index()
```

```python
# create a new csv 
cleaned_covid_year_df.to_csv("../Output/Cleaned Datasets/EDD_cleaned_dataset.csv", index=False)
```
